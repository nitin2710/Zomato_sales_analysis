```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Zomato.csv")
df
```

```python
def divide(value):
    value = str(value).split("/")
    value = value[0]
    return float(value)

df['rate'] = df['rate'].apply(divide)
print(df)
```

```python
df.info()
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("Type of restuarent")
plt.ylabel("Count of restuarents")
```

```python
grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result, c='Green' , marker = 'o')
plt.xlabel("Type of restuarents",c = 'Red' , size = 20)
plt.ylabel("No of Votes",c = 'Red' , size = 20)
```

```python
plt.hist(df['rate'],bins = 5)
plt.title("Rating of Restaurants")
plt.show()
```

```python
couple_order=df['approx_cost(for two people)']
sns.countplot(x=couple_order)
```

```python
sns.boxplot(x = 'online_order' , y = 'rate' , data = df)
```

```python
pivot_table = df.pivot_table(index = 'listed_in(type)' , columns = 'online_order' , aggfunc = 'size' , fill_value = 0)
sns.heatmap(pivot_table ,annot = True, cmap = 'YlGnBu' , fmt = 'd')
plt.title("Heatmap")
```
